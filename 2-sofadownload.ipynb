```python
import pysofaconventions
import sofadownloadutils
```

```python
# fetch the links from a directory and show the first five...

# ".sofa" only saves links for .sofa files
# "10" only saves the first 10 links

theLinks = sofadownloadutils.fetch_links("http://sofacoustics.org/data/database/ari/", ".sofa", 10)
theLinks[:5]
```

```python
# ...then download them to the /database/ari/ subdirectory in the folder that this code is run.

# To download all the files straight to the directory (kinda messy), just leave that field with blank quotes ''.

# By default, this only prints a message once before downloading, and again after all of the files are downloaded.
# So if you're downloading a huge dataset, you won't get a message for a while.
# If you'd like to get a message at each file, just set printFileOutput=True .

sofadownloadutils.download_links(theLinks, 'datasets/ari/', printFileOutput=False)
```
